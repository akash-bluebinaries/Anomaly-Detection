```python
import os,sys
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('D:/bmsAnomalyDetection/application/Anomaly-detection/notebook/data/23-07-24.csv')
pd.set_option('display.max_columns', None)
df.head()
```

```python
df['BMS_max_cell_temp_id'] = df['BMS_max_cell_temp_id'].astype('object')
df['BMS_state'] = df['BMS_state'].astype('object')
df['BMS_min_cell_temp_id'] = df['BMS_min_cell_temp_id'].astype('object')
df['BMS_max_cell_voltage_id'] = df['BMS_max_cell_voltage_id'].astype('object')
df['BMS_min_cell_voltage_id'] = df['BMS_min_cell_voltage_id'].astype('object')
df['OBC_mux'] = df['OBC_mux'].astype('object')
df['OBC_port_status'] = df['OBC_port_status'].astype('object')
df['OBC_overvoltage_fault'] = df['OBC_overvoltage_fault'].astype('object')
df['OBC_overcurrent_fault'] = df['OBC_overcurrent_fault'].astype('object')
df['OBC_port_weld_fault'] = df['OBC_port_weld_fault'].astype('object')


df = df.drop(['timestamps'], axis = 1)
```

```python
import joblib

# Load the preprocessor
loaded_preprocessor = joblib.load('D:/bmsAnomalyDetection/application/Anomaly-detection/notebook/artifacts/Preprocessor.pkl')

# Load the model
loaded_model = joblib.load('D:/bmsAnomalyDetection/application/Anomaly-detection/notebook/artifacts/Model.pkl')
```

```python
processedData= pd.DataFrame(loaded_preprocessor.fit_transform(df), columns=df.columns)
```

```python
processedData['prediction'] = loaded_model.predict(processedData)
```
